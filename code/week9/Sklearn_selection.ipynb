# RandomForest + Scikit Learn
----
* 데이터 셋이 작기에 Train/Test set으로만 나눠서 진행
* Training : 75%, Test : 25%

```python
import pandas as pd

df = pd.read_csv("../../COALA_DS_DATA/COALA_DS_DATA/week4/data/iris.csv")
df
```

```python
input_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
target_col = ['Species']

input_data = df[input_cols]
target_data = df[target_col]
```

```python
from sklearn.model_selection import train_test_split

# 자동으로 75%, 25%를 나눠줌
# 고르게 나눠주기까지 한다.
x_train, x_test, y_train, y_test = train_test_split(input_data, target_data)

x_train
```

```python
x_test.count()
```

## DecisionTree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier()
tree.fit(x_train, y_train)

print('training set accuracy :', tree.score(x_train, y_train))
print('training set accuracy :', tree.score(x_test, y_test))
```

```python
prediction = tree.predict(x_test)
prediction
```

```python
x_test['Species'] = prediction

x_test = x_test.sort_index()
y_test = y_test.sort_index()
print(y_test)
print(x_test)
#print(x_test['Species'], y_test)
#print(x_test['Species'] == y_test)
#print(y_test['Species'])
#print(x_test == y_test['Species'])
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators = 1000) 
# n_estimators : randomforest안에 몇개의 DC트리를 생성할 것인가
# default 100개
forest.fit(x_train, y_train.values.ravel())

#print(y_train.values.ravel())
print('training set accuracy :', forest.score(x_train, y_train))
print('test set accuracy :', forest.score(x_test, y_test))

prediction_by_forest = forest.predict(x_test)
prediction_by_forest
```
