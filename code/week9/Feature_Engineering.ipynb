# Feature Engineering
-----
대상의 특징 정보를 바탕으로 예측

```python
import pandas as pd

df = pd.read_csv("../../COALA_DS_DATA/COALA_DS_DATA/week4/data/iris.csv")
df.head(10)
```

```python
df.info() # 데이터 프레임의 정보들
```

```python
df.describe() # 통계 정보
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set()
```

### 꽃받침의 길이, 너비

```python
sns.catplot(data = df, x = 'SepalLengthCm', y = 'SepalWidthCm', hue = "Species", aspect = 2)
```

### 꽃잎의 길이, 너비

```python
sns.catplot(data = df, x = 'PetalLengthCm', y = 'PetalWidthCm', hue = "Species", aspect = 2)
```

### 꽃받침 길이에 따른 밀집도

```python
facet = sns.FacetGrid(df, hue = "Species", aspect = 2)
facet.map(sns.kdeplot, 'SepalLengthCm')
facet.add_legend()
plt.show()
```

## Box plot
---
* **유효한 범위를 알 수 있음**
* 하지만 **밀집도는 알 수 없다**

```python
sns.catplot(kind = "box", data = df, x = 'Species', y = 'SepalLengthCm', aspect = 2)
```

## Violin Plot
---
**값의 범위와 밀집정도를 동시에 시각적으로 확인 가능**

### kdeplot 과 box plot의 장점을 합쳐놓은 plot
---

### Violin plot (x : 종, y : 꽃받침 길이)

```python
sns.violinplot(data = df, x = 'Species', y = 'SepalLengthCm', aspect = 2)
```

### Violin plot (x : 종, y : 꽃받침 너비)


```python
sns.violinplot(data = df, x = 'Species', y = 'SepalWidthCm', aspect = 2)
```

### Violin plot (x : 종, y : 꽃잎 길이)

```python
sns.violinplot(data = df, x = 'Species', y = 'PetalLengthCm', aspect = 2)
```

### Violin plot (x : 종, y : 꽃잎 너비)

```python
sns.violinplot(data = df, x = 'Species', y = 'PetalWidthCm', aspect = 2)
```
