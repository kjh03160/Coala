## Feature Scaling을 적용한 데이터의 Visualization

```python
import pandas as pd

df = pd.read_csv("../../COALA_DS_DATA/COALA_DS_DATA/week5/data/house.csv")

# 필요 없는 특징 없애기
house_data = df.drop(['id', 'date'], axis = 1)    # aixs=1(columns)은 열을 따라 동작
house_data

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_data = scaler.fit_transform(house_data) # 반환값 -> ndarray
scaled_data = pd.DataFrame(scaled_data, columns = house_data.columns)

scaled_data
```

```python
plt.figure(figsize = (20, 15))
sns.heatmap(scaled_data.corr(), annot = True, fmt = '.2f', square = True)
plt.show()
```

```python
scaled_data.corr() # 각 열에 관한 피어슨 상관계수 값들 보여줌
```

```python
# 산포도 시각화
sns.catplot(data = scaled_data, x = 'bathrooms', y = 'price', aspect = 2)
```
