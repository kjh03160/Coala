```python
plt.plot(['Math', 'Korean', 'English'], [90, 100 ,50])
plt.plot(['Math', 'Korean', 'English'], [70, 60 ,90])

plt.xlabel('Subjects')
plt.ylabel('Score')
plt.title('Midterm')


plt.legend(['Kim', 'Choi'])
plt.show()
```

```python
x = ['Math', 'Korean', 'English']
y = [90, 100, 50]

plt.xlabel('Subjects')
plt.ylabel('Score')

plt.bar(x, y, width = 0.1, color="black")
plt.show()
```

```python
plt.plot([1, 2, 3], [90, 100, 50], marker='+', markersize = 30, linestyle='--')

plt.grid(True)
plt.axis([1, 3, 80, 100])

plt.show()
```

# 판다스로 csv 데이터 불러오기

```python
import pandas as pd
```

```python
train = pd.read_csv('../../COALA_DS_DATA/COALA_DS_DATA/week2/data/train.csv')

train
```

```python
train.head() # 상위 5개
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
sns.relplot(data = train, x = 'Age', y= 'Survived')
```

```python
sns.relplot(data = train, x = 'Age', y= 'Survived', hue = 'Sex', aspect = 3) # aspect는 가로/세로 비율
```

```python
sns.catplot(data = train, x = 'Pclass', y = 'Fare', hue="Sex", aspect = 3)
```

```python
sns.catplot(data = train, x = 'Pclass', y = 'Fare', aspect = 3, jitter = False)
```

```python
sns.pairplot(data = train, hue="Sex",
            x_vars = ['Pclass', 'Age', 'Fare'],
            y_vars = ['Survived', 'Age', 'Fare'], height = 3)
```

```python
sns.catplot(data = train, x = 'Pclass', y = 'Fare', hue = "Sex", kind= 'box')
```
