### 타이타닉의 사망자, 생존자를 Bar Chart로 표현해보세요.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

train = pd.read_csv('../../COALA_DS_DATA/COALA_DS_DATA/week2/data/train.csv')
train
```

```python
# df.loc[조건, 열]
not_survived = train.loc[train["Survived"] == 0, 'PassengerId'].count()
survived = train.loc[train["Survived"] == 1, 'PassengerId'].count()

plt.bar(["0"], [not_survived], color = 'blue')
plt.bar(["1"], [survived], color = 'red')

plt.xlabel("Survived")
plt.ylabel("count")
plt.legend(["Dead", "Survived"])  # 범례 표시

plt.show()
```
